## Codificación de variables categóricas:

vamos a codificar todas nuestras variables categoricas, en particular el precio, el mantenimiento, el tamaño del maletero, la seguridad y nuestra variable objetivo son variables ordinales donde le asignare un valor desde 1 como el mas pequeño o deficiente y hasta el dato mas grande o eficiente que poseea el campo

```python
# defino un diccionario con valores a remplazar 
r_buying={"low":1,"med":2,"high":3,"vhigh":4}
r_maint={"low":1,"med":2,"high":3,"vhigh":4}
r_lug_boot={"small":1,"med":2,"big":3}
r_safety={"low":1,"med":2,"high":3}
r_class={"unacc":1,"acc":2,"good":3,"vgood":4}
```

```python
# aplicamos el reemplazo con nustro diccionario creado correspondiente
df['buying'] = df['buying'].replace(r_buying)
df['maint'] = df['maint'].replace(r_maint)
df['lug_boot'] = df['lug_boot'].replace(r_lug_boot)
df['safety'] = df['safety'].replace(r_safety)
df['class'] = df['class'].replace(r_class)

df
```

Para los campos de numero de puertas y numero de personas transportar, se remplaza el valor categorico por un valor nuemro al final indica que puede llevar a mas personas o que tiene mas puertas, en mi caso dejare todo lo que sea mayor a 5 puertas como 5 y en personas se reemplaza por 5

```python
# reemplazo valor en doors
df['doors'] = df['doors'].replace('5more', '5').astype(int)
# reemplazo valor en persons
df['persons'] = df['persons'].replace('more', '5').astype(int)
df.info()
```

Con este proceso ya podemos empezar a correr los modelos estudiar

## Implementación y evaluación

Se avaluara tres modelos para estudiar su metricas y determinar cual puede ser mejor con los datos a trabajar 

### 1. Arboles de decisión
primero vamos a cargar librerias y posterior vasmos a separar nuestro dataset en train y test

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score,recall_score,precision_recall_curve
from sklearn.tree import plot_tree
```

```python
X=df.drop(columns='class')
y=df['class']
X
```

```python
# Dividir el conjunto de datos y entrenar el modelo
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
## Instancio el modelo
modelo = DecisionTreeClassifier(max_depth=10, random_state=42)
```

```python
## Entreno el modelo
modelo.fit(X_train, y_train)
```

```python
# Evaluar el modelo
y_pred = modelo.predict(X_test)
print(f"Exactitud del modelo: {accuracy_score(y_test, y_pred):.2f}")
```

```python
X_test.iloc[0]
```

```python
modelo.predict(X_test.iloc[0].values.reshape(1, -1))
```

```python
plt.figure(figsize=(40,20))
plot_tree(modelo, filled=True, feature_names=X.columns, class_names=["1","2","3","4"], rounded=True, proportion=True)
plt.savefig("Arbol_decision.png")
#plt.show()
```

```python
# Evaluar el modelo
y_pred = modelo.predict(X_test)
print(f"Reporte del modelo arbol de decision:\n {classification_report(y_test, y_pred)}")
```

```python
# Evaluar el modelo
y_pred = modelo.predict(X_test)
print(f"Matriz de confusión:\n {confusion_matrix(y_test, y_pred)}")
```

Observamos que modelo es bastante bueno, a nivel de presicion para cada una de las clases, es bastante capas de predecir, en particular para la clase 1 (inaceptable) y clase 2 (aceptable) tuvo una presicion alta. la medida general del modelo esta bastante alta con un 0,97 y el f1 demuetra que apesar el desbalanceo de los datos el  modelo tiene un buen equilibrio entrela precision y el recall.

```python
# Evaluar el modelo
y_pred = modelo.predict(X_train)
print(f"Reporte del modelo arbol de decision:\n {classification_report(y_train, y_pred)}")
```

```python
from sklearn.model_selection import learning_curve
 
train_sizes, train_scores, test_scores = learning_curve(estimator=modelo,X=X_train,y=y_train,train_sizes=np.linspace(0.1, 1.0, 10),cv=10,n_jobs=-1)
 
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
 
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training accuracy')
plt.plot(train_sizes, test_mean, color='orange', linestyle='--', marker='s', markersize=5, label='Validation accuracy')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```

Vemos que medida que el los datos de entramiento aumentan tambien aumenta la exactitud del modelo.
